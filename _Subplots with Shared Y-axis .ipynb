```python
import matplotlib.pyplot as plt
days = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
hours_studied = [2, 3, 4, 3, 5, 1, 0]
hours_slept = [7, 6, 8, 7, 6, 8, 9]
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 5))
ax1.plot(days, hours_studied, marker='o', color='green', label='Hours Studied')
ax1.set_title('Hours Studied Over a Week')
ax1.set_xlabel('Day of the Week')
ax1.set_ylabel('Number of Hours')
ax2.plot(days, hours_slept, marker='o', color='red', label='Hours Slept')
ax2.set_title('Hours Slept Over a Week')
ax2.set_xlabel('Day of the Week')
fig.text(0.04, 0.5, 'Number of Hours', va='center', rotation='vertical')
plt.tight_layout(rect=[0.05, 0, 1, 1])
plt.show()
```
